```python
#Q:3 Compare and contrast mutable and immutable objects in Python with examples? 
# Mutable Objects:
#- Can Change: You can modify them.
#- Examples: Lists, dictionaries.

#python
# List example
my_list = [1, 2, 3]
my_list[0] = 10  # Changing an element
print(my_list)  # Output: [10, 2, 3]

# Dictionary example
my_dict = {'a': 1, 'b': 2}
my_dict['a'] = 10  # Changing a value
print(my_dict)  # Output: {'a': 10, 'b': 2}

# Immutable Objects:
#- Cannot Change: You cannot modify them.
#- Examples: Strings, tuples.

#python
# String example
my_string = "hello"
my_string[0] = 'H'  # This will cause an error

# Tuple example
my_tuple = (1, 2, 3)
my_tuple[0] = 10  # This will cause an error
```

```python
#Q:4 Discuss the different types of operators in Python and provide examples of how they are used.

# Types of Operators in Python:


# Arithmetic Operators:

# Perform basic arithmetic operations.
# Examples: +, -, *, /, //, %, **.
a = 10
b = 3
print(a + b)   # Addition: 13
print(a - b)   # Subtraction: 7
print(a * b)   # Multiplication: 30
print(a / b)   # Division: 3.333
print(a // b)  # Floor Division: 3
print(a % b)   # Modulus: 1
print(a ** b)  # Exponentiation: 1000

# Comparison Operators:

# Compare values and return a boolean result.
# Examples: ==, !=, >, <, >=, <=.
x = 5
y = 10
print(x == y)  # Equal to: False
print(x != y)  # Not equal to: True
print(x > y)   # Greater than: False
print(x < y)   # Less than: True
print(x >= y)  # Greater than or equal to: False
print(x <= y)  # Less than or equal to: True

# Logical Operators:

# Perform logical operations and return a boolean result.
# Examples: and, or, not.

t = True
u = False
print(t and u)  # Logical AND: False
print(t or u)   # Logical OR: True
print(not t)    # Logical NOT: False

# Assignment Operators:

# Assign values to variables.
# Examples: =, +=, -=, *=, /=, //=, %=, **=.

p = 5
p += 2  # Equivalent to: x = x + 2
print(p)  # Output: 7

q = 10
q *= 3  # Equivalent to: y = y * 3
print(q)  # Output: 30

# Bitwise Operators:

# Perform bit-level operations on integers.
# Examples: &, |, ^, ~, <<, >>.

f = 6  # Binary: 110
g = 3  # Binary: 011
print(f & g)  # Bitwise AND: 2 (Binary: 010)
print(f | g)  # Bitwise OR: 7 (Binary: 111)
print(f ^ g)  # Bitwise XOR: 5 (Binary: 101)
print(~f)     # Bitwise NOT: -7
print(f << 1) # Left Shift: 12 (Binary: 1100)
print(f >> 1) # Right Shift: 3 (Binary: 011)


# Identity Operators:

# Compare the memory locations of two objects.
# Examples: is, is not.
l = [1, 2, 3]
m = l
n = [1, 2, 3]
print(l is m)      # True
print(l is not n)  # True
```

```python
#Q:5 Explain the concept of type casting in Python with examples?

# Type casting is the process of converting a variable from one data type to another.
# Integer to Float
x = 10
y = float(x)
print(y)  # Output: 10.0

# String to Integer
s = "123"
n = int(s)
print(n)  # Output: 123

# List to Tuple
my_list = [1, 2, 3]
my_tuple = tuple(my_list)
print(my_tuple)  # Output: (1, 2, 3)
```

```python
#Q:6 How do conditional statements work in Python Illustrate with examples?

# Conditional Statements in Python:
# Conditional statements allow you to execute certain blocks of code based on whether a condition is true or false.
#Syntax
# if condition:
    # block of code to execute if the condition is true
# elif condition:
    # block of code to execute if the previous condition was false and this condition is true
# else:
    # block of code to execute if all above conditions are false
age = 25
if age >= 18 and age < 65:
    print("You are eligible to work")
elif age >= 65:
    print("You are eligible for retirement benefits")
else:
    print("You are not eligible to work yet")
```

```python
#Q:7 Describe the different types of loops in Python and their use cases with examples.

# Types of Loops in Python:
# For Loop:

# Iterates over a sequence (such as a list, tuple, string) or other iterable objects.
# Example: Iterating over a list.
# Use case: When you know the number of iterations beforehand or want to iterate over a sequence.
# Iterate over a list
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    print(num)
    
    
# While Loop:

# Repeats a block of code as long as a condition is true.
# Example: Summing numbers until a condition is met.
# Use case: When you don't know beforehand how many times the loop will execute, but it depends on a condition.
# Sum of numbers until the sum exceeds 10
total = 0
num = 1
while total <= 10:
    total += num
    num += 1
print("Sum exceeded 10. Total:", total)
```
